```python

import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
# ============================================================
# 1. DATA (any text goes here)
# ============================================================
 
text = "the quick brown fox jumps over the lazy dog"
 
chars = sorted(list(set(text)))
char_to_idx = {c: i for i, c in enumerate(chars)}
idx_to_char = {i: c for i, c in enumerate(chars)}
vocab_size = len(chars)
 
print(f"Text: {text}")
print(f"Vocab size: {vocab_size} unique characters\n")
 
```

```python
# input = all characters except last, target = all characters except first
input_seq = text[:-1]
target_seq = text[1:]
 
input_idx = [char_to_idx[c] for c in input_seq]
target_idx = [char_to_idx[c] for c in target_seq]
 
X = tf.one_hot(input_idx, vocab_size)
X = tf.expand_dims(X, axis=0)          # shape: (1, seq_len, vocab_size)
y = tf.expand_dims(target_idx, axis=0)  # shape: (1, seq_len)
 
```

```python
# ============================================================
# 2. MODEL - plain vanilla RNN only
# ============================================================
 
model = models.Sequential([
    layers.Input(shape=(None, vocab_size)),
    layers.SimpleRNN(32, return_sequences=True),   # <-- the RNN, nothing else
    layers.Dense(vocab_size)
])
 
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```
